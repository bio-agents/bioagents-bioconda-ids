# overlap study for bioconda and bioagents

## get list of bioagents entries

```python
import requests
bioagents_list = []
next_bioagents_page = '?page=1'
while next_bioagents_page is not None:
    print('getting page %s' % next_bioagents_page)
    page = requests.get('https://bio.agents/api/agent/' + next_bioagents_page).json()
    bioagents_list += page['list']
    next_bioagents_page = page['next']
```

_and get the list of bioagents ids for these entries_

```python
from pandas import DataFrame
bioagents_ids = [element['id'] for element in bioagents_list]
bioagents_ids_processed = {}
for b in bioagents_list:
    if 'BioConductor' in b['collectionID']:
        bioagents_ids_processed['BIOCONDUCTOR-%s' % b['id'].upper()] = b['id']
    if 'EMBOSS' in b['collectionID']:
        bioagents_ids_processed['BIOCONDUCTOR-%s' % b['id'].upper()] = b['id']
    else:
        bioagents_ids_processed[b['id'].upper()] = b['id']
bioagents_ids = DataFrame(list(bioagents_ids_processed.items()), columns=['common', 'bioagents'])
bioagents_ids = bioagents_ids.set_index('common')
```

## get list of bioconda entries

```python
#repodata_url = 'https://conda.anaconda.org/bioconda/linux-64/repodata.json'
#repodata = requests.get(repodata_url).json()
!rm repodata.json && wget https://conda.anaconda.org/bioconda/linux-64/repodata.json
import json
bioconda_list = json.load(open('repodata.json'))
```

```python
print(len(bioconda_list['packages'].items()))
```

_and get the list of bioconda package names_

```python
#bioconda_ids = [element['name'] for element in bioconda_list]
#print(bioconda_ids))
bioconda_ids = set([package['name'] for key, package in bioconda_list['packages'].items()])
bioconda_ids_processed = {}
for bioconda_id in bioconda_ids:
    if len(bioconda_id.split('-'))>1 and bioconda_id.split('-')[0]!='bioconductor':
        bioconda_ids_processed[bioconda_id.split('-')[1].upper()] = bioconda_id
    else:
        bioconda_ids_processed[bioconda_id.upper()] = bioconda_id
bioconda_ids = DataFrame(list(bioconda_ids_processed.items()), columns=['common', 'bioconda'])
bioconda_ids = bioconda_ids.set_index('common')
print(bioagents_ids)
print(bioconda_ids)
```

this is the intersection of bioagents/biocondas entries, based on strict name equality

```python
bioconda_ids.join(bioagents_ids).to_csv('mapping.csv',na_rep='null')
bioconda_ids.join(bioagents_ids).dropna().to_csv('mapping_matchonly.csv',na_rep='null')
```
